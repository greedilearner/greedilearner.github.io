```python
import tensorflow as tf
```

```python
fashion_minst=tf.keras.datasets.california_housing.load_data()
(X_train_full,y_train_full),(X_test,y_test)=fashion_minst
X_train,y_train =X_train_full[:-100],y_train_full[:-100]
X_valid,y_valid=X_train_full[-100:],y_train_full[-100:]
```

```python
tf.random.set_seed(42)
norm_layer= tf.keras.layers.Normalization(input_shape=X_train.shape[1:])
model = tf.keras.Sequential([
    norm_layer,
    tf.keras.layers.Dense(50,activation='relu'),
    tf.keras.layers.Dense(50,activation='relu'),
    tf.keras.layers.Dense(50,activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
model.compile(loss='mse',
              optimizer=optimizer,
              metrics=["RootMeanSquaredError"])
```

```python
norm_layer.adapt(X_train)
history=model.fit(X_train,y_train,epochs=30,validation_data=(X_valid,y_valid))
```

```python
mse_test,rmse_test=model.evaluate(X_test,y_test)
X_new =X_test[:3]
y_pred=model.predict(X_new)
```
