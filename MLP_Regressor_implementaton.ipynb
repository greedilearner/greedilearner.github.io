**Implementation of Mutli-layer Perceptron**

```python
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

```

```python
housing = fetch_california_housing()
```

```python
Xfull_train,x_test,yfull_train,y_test =train_test_split(housing.data,housing.target,random_state=42)
x_train,x_vaild,y_train,y_vaild=train_test_split(Xfull_train,yfull_train,random_state=42)
```

```python
mlp_reg=MLPRegressor(hidden_layer_sizes=[5,5,5],random_state=42)
Pipeline=make_pipeline(StandardScaler(),mlp_reg)
```

```python
Pipeline.fit(x_train,y_train)
```

```python
y_pred=Pipeline.predict(x_vaild)
```

```python
import sklearn
print (sklearn.__version__)
```

```python
rmes=mean_squared_error(y_vaild,y_pred)
```

```python
print(rmes)
```

```python
import numpy as np
rmes = np.sqrt(rmes)
```
